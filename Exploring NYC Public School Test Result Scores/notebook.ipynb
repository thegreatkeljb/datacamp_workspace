Every year, American high school students take SATs, which are standardized tests intended to measure literacy, numeracy, and writing skills. There are three sections - reading, math, and writing, each with a **maximum score of 800 points**. These tests are extremely important for students and colleges, as they play a pivotal role in the admissions process.

Analyzing the performance of schools is important for a variety of stakeholders, including policy and education professionals, researchers, government, and even parents considering which school their children should attend. 

You have been provided with a dataset called `schools.csv`, which is previewed below.

You have been tasked with answering three key questions about New York City (NYC) public school SAT performance.

```python
# Re-run this cell 
import pandas as pd

# Read in the data
schools = pd.read_csv("schools.csv")

# Preview the data
schools.head()

# Start coding here...
# Add as many cells as you like...
```

```python
#Check for dup school. If there's 0 in school_name, group the data. Otherwise, proceed.
schools['school_name'].value_counts(sort=True).head()
```

```python
schools.isna().any() #The initial dataset was missing values in percent_tested column
```

```python
#Prep the data
schools = schools.fillna(0) #Fill the missing value
schools.head() #Preview the data after preprocessing
```

```python
#NYC schools with best math results

eighty_percent = 800 * 0.80 # 80% of maximum score possible (800)

best_math_schools = schools.loc[schools['average_math'] >= eighty_percent, ['school_name', 'average_math']]
best_math_schools = best_math_schools.sort_values('average_math', ascending=False)
best_math_schools.head() #Preview the best math schools
```

```python
#Get the top 10 schools based on the combined SAT scores

schools['total_SAT'] = schools[['average_math', 'average_reading', 'average_writing']].sum(axis=1)

top_10_schools = schools[['school_name', 'total_SAT']].sort_values('total_SAT', ascending=False)

top_10_schools = top_10_schools.iloc[0:10] 

top_10_schools #Displays top 10 schools based on SAT Scores
```

```python
#Borough with largest std dev in the combined SAT score

largest_std_dev = schools.groupby('borough')['total_SAT'].agg(['std', 'mean']).round(2).sort_values('std', ascending=False)

largest_std_dev = largest_std_dev.iloc[:1] #Only take one row, borough w/ largest std dev

largest_std_dev['num_schools'] = schools['borough'].value_counts(sort=True)

#Rename columns to match the requirements
largest_std_dev = largest_std_dev.rename(columns={'std':'std_SAT',
                               'mean':'average_SAT'})
```
