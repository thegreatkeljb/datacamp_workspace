```python
# Importing pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt

# Read in the Netflix CSV as a DataFrame
netflix_df = pd.read_csv("netflix_data.csv")
```

```python
print(netflix_df.iloc[0]) #Checking the structure of data entries
```

```python
import numpy as np

#Get movies released 1990 - 1999
films_decade = netflix_df[np.logical_and(netflix_df['release_year'] >= 1990, netflix_df['release_year'] <= 1999)]
movie_1990s = films_decade[films_decade['type'] == 'Movie']

#Most frequent duration in 90s movies
duration = int(np.median(movie_1990s['duration'])) 

#Count the number of short action movies
shortmovies = movie_1990s[np.logical_and(movie_1990s['duration'] < 90, movie_1990s['genre'] == 'Action')]
short_movie_count = len(movie_1990s[np.logical_and(movie_1990s['duration'] < 90, movie_1990s['genre'] == 'Action')])
```

```python
print(f"Most frequent duration in 90s movies: {duration}")
print(f"Short action movies count in 90s: {short_movie_count}")
```
