The Nobel Prize has been among the most prestigious international awards since 1901. Each year, awards are bestowed in chemistry, literature, physics, physiology or medicine, economics, and peace. In addition to the honor, prestige, and substantial prize money, the recipient also gets a gold medal with an image of Alfred Nobel (1833 - 1896), who established the prize.

![](Nobel_Prize.png)

The Nobel Foundation has made a dataset available of all prize winners from the outset of the awards from 1901 to 2023. The dataset used in this project is from the Nobel Prize API and is available in the `nobel.csv` file in the `data` folder.

In this project, you'll get a chance to explore and answer several questions related to this prizewinning data. And we encourage you then to explore further questions that you're interested in!

```python
# Loading in required libraries
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
#Inspect the dataset
df = pd.read_csv('data/nobel.csv')
print(f"Dataset shape: {df.shape}\n")
df.info()
df.head()
```

```python
#Most commonly awarded gender and birth
top_gender, top_country = df[['sex', 'birth_country']].value_counts().idxmax()

print(f"Most awarded country: {top_country}\nMost awarded gender: {top_gender}")
```

```python
#Decade each winner was born
df['decade'] = (df['year'] // 10) * 10

df[['birth_country', 'year', 'decade']].head()
```

```python
#Decade with highest ratio of US-born Nobel Prize winner
winners_per_decade = df.groupby('decade').size()
usa_winners_per_decade = df[df['birth_country'] == 'United States of America'].groupby('decade').size()
ratio_per_decade =  usa_winners_per_decade/ winners_per_decade
max_decade_usa = int(ratio_per_decade.idxmax())

print(f"Decade with highest ratio of US-born Nobel Prize winner: {max_decade_usa}")
```

```python
#Highest proportion of female laureates in decade and Nobel Prize category combination
total_combination = df.groupby(['decade', 'category']).size()
female_combination = df[df['sex'] == 'Female'].groupby(['decade', 'category']).size()
female_proportion = female_combination / total_combination
key, val = female_proportion.idxmax()
key = int(key)
max_female_dict = {
    key:val
}

print(f"Dictionary with highest proportion of female winners: {max_female_dict}")
```

```python
#First woman to receive a nobel prize
first_woman = df[df['sex'] == 'Female'].sort_values('year').iloc[0]
first_woman_name, first_woman_category = first_woman[['full_name', 'category']]

print(f"The first woman to receive a nobel prize is {first_woman_name} in {first_woman_category} category")
```

```python
#Individuals or organizations won more than once
times_won = df[['full_name']].value_counts().reset_index(name='times_won')

twice_or_more = times_won[times_won['times_won'] > 1]
repeat_list = list(twice_or_more['full_name'])
print(f"List of individuals or organizations won more than once:\n{repeat_list}")
```
